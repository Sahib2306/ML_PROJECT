```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)

n = 50

data = {
    "StudentID": range(1, n+1),
    "Age": np.random.randint(17, 22, n),
    "StudyHours": np.round(np.random.uniform(0.5, 5, n), 1),
    "Attendance": np.random.randint(50, 100, n),
    "Assignments": np.random.randint(1, 10, n),
    "MidtermMarks": np.random.randint(20, 80, n),
    "FinalMarks": np.random.randint(25, 90, n),
    "ParentalSupport": np.random.randint(0, 2, n),
    "InternetAccess": np.random.randint(0, 2, n),
}


df = pd.DataFrame(data)

```

```python
df["Pass"] = ((df["MidtermMarks"] + df["FinalMarks"]) / 2 > 50).astype(int)
```

```python
df.to_csv("student_data.csv", index=False)

print("Dataset generated successfully!")
```
